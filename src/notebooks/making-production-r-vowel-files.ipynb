```python
import pandas as pd
import re
```

```python
df = pd.read_csv('data/NIKL_lC+Vowel.csv')
df.head()
```

```python
morphology_r_list = df['Morphology_R']
```

```python
ur = df['UR']
sr = df['SR']
```

```python
dot_sr = [s[0] + '.' + s[1:] for s in sr]
```

```python
production_r_list = []
for m, u, s in zip(morphology_r_list, ur, dot_sr):
    new_m = re.sub(u, s, m)
    production_r_list.append(new_m)
```

```python
print(len(morphology_r_list), len(ur), len(dot_sr), len(production_r_list))
```

```python
df['Production_R'] = production_r_list
```

```python
df.head()
```

```python
df.to_csv('data/modified_NIKL_IC+Vowel.csv', index = False)
```
