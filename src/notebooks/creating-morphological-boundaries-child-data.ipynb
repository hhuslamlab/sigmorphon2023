```python
import pandas as pd
import re
from collections import Counter
```

```python
df = pd.read_csv('../data/Ko_complete.csv')
df.head()
```

```python
morphology_r_list = df['Morphology_R']
```

```python
ur = df['UR']
```

```python
patterns = Counter(ur)
```

```python
keys = []
for pat in patterns:
    k = pat.split('-')[0]
    keys.append(k)
```

```python
pat_dict = dict(zip(keys, patterns))
```

```python
print(len(keys))
```

```python
print(len(patterns))
```

```python
for morph in morphology_r_list:
    for key in pat_dict:
        morph_bdry = re.sub(key, pat_dict[key], morph)
        print(morph_bdry)
```
